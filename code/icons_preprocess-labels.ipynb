# Add New Labels to Ground Truth

```python
new_labels_df = pd.read_json("../data/labeling/labels_clean.json")
```

```python
new_ids = new_labels_df["sample_id"].to_list()
new_labels = new_labels_df["label"].to_list()
```

```python
new_descrip = {}
for id, label in zip(new_ids, new_labels):
    if id not in new_descrip:
        new_descrip[id] = []
    new_descrip[id].append(label)
```

```python
len(new_descrip)
```

### Zero-shot Ablation

```python
from model import Icon, Dataset
```

```python
zero_shot = Dataset("")
zero_shot.load_pkl("ablation_zero-shot.pkl")
```

```python
for icon in zero_shot.icons:
    for k, v in icon.descriptions.items():
        icon.descriptions[k] = [v]
    icon.descriptions["ground-truth"].extend(new_descrip[str(icon.dhash)])
```

```python
zero_shot.save_pkl("ablation_zero-shot.pkl")
zero_shot.save_json("../outputs/zero-shot.json")
```

### k-shot Single Generation

```python
train = Dataset("")
train.load_pkl("k-shot_train.pkl")

test = Dataset("")
test.load_pkl("k-shot_test.pkl")
```

```python
print(len(train.icons), len(test.icons))
```

```python
for d in [train, test]:
    for icon in d.icons:
        for k, v in icon.descriptions.items():
            icon.descriptions[k] = [v]
        icon.descriptions["ground-truth"].extend(new_descrip[str(icon.dhash)])
```

```python
train.save_pkl("k-shot_train.pkl")
train.save_json("../outputs/k-shot_train.json")

test.save_pkl("k-shot_test.pkl")
test.save_json("../outputs/k-shot_test.json")
```
